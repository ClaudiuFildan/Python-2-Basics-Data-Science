```python
import numpy as np
x=np.arange(0,100)
y=x*2
z=x**2
```

```python
# Import matplot.pyplot as plt and set %matplotlib inline if you are using the jupiter
# notebook. What command do you use if you are not using the jupyter notebook?
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.show()
```

```python
# Exercise 1
# Createa figure object called fig using plt.figure()
# Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axisax.
# Plot (x,y) on that axes and set the labels and titles to match the plot below:
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

```python
# Exercise 2
# Create a figure object and put two axes on it, ax1 and ax2. Locate at [0,0,1,1] and 
# [0.2,0.5,.2,.2] respectively.
# Now plot (x,y) on both axes. And call your figure object to show it.
```

```python
fig=plt.figure()

ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,0.2,0.2])

ax1.plot(x,y,color='red')
ax2.plot(x,y,color='red')
```

```python
# Exercise 3 
# Create the plot below by adding twoaxes to a figure object at [0,0,1,1] and
# [0.2,0.5,.4,.4]
# Now use x,y and z arrays to recreate the plot below. Notice the x limits and 
# y limits on the inserted plot:
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,0.4,0.4])

# Large
ax.plot(x,y)
ax.set_xlabel('X')
ax.set_ylabel('Z')

# Insert
ax2.plot(x,y)
ax2.set_title('zoom')
ax2.set_xlabel('x')
ax2.set_xlabel('y')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```
