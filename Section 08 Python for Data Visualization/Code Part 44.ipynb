```python
x=np.linspace(0,5,11)
y=x**2
```

```python
# FUNCTIONAL 
plt.plot(x,y,'r--')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# 00
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=3,alpha=0.5,linestyle='--') # RGB HEX
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=1,ls='-',marker='o',markersize=20,markerfacecolor='yellow',markeredgewidth=3)

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
